## Importing libraries

```python
import re
import math
from collections import defaultdict, Counter
from dataclasses import dataclass
from typing import List, Dict, Set, Tuple, Union

from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

## Loading Corpus

```python
def load_corpus(limit: int = 2000):
    """
    Load a corpus of raw text documents.
    limit can be >= 500 to satisfy the assignment.
    """
    dataset = fetch_20newsgroups(
        subset="all",
        remove=("headers", "footers", "quotes")
    )
    docs = dataset.data[:limit]
    return docs
```

```python
## 3. NORMALIZATION
```

```python
STOPWORDS = set(ENGLISH_STOP_WORDS)

def normalize_text(text: str) -> List[str]:
    """
    Lowercase, extract alphabetic tokens, remove stopwords.
    """
    tokens = re.findall(r"[a-zA-Z]+", text.lower())
    tokens = [t for t in tokens if t not in STOPWORDS]
    return tokens
```

## 4. BUILD INVERTED INDEX

```python
def build_inverted_index(documents: List[str]):
    """
    Returns:
      index: term -> {doc_id: tf}
      doc_tokens: doc_id -> Counter(term frequencies)
      normalized_docs: doc_id -> list of normalized tokens
    """
    index = defaultdict(dict)
    doc_tokens = {}
    normalized_docs = {}

    for doc_id, doc in enumerate(documents):
        tokens = normalize_text(doc)
        normalized_docs[doc_id] = tokens
        tf_counter = Counter(tokens)
        doc_tokens[doc_id] = tf_counter

        for term, freq in tf_counter.items():
            index[term][doc_id] = freq

    return index, doc_tokens, normalized_docs
```

##  5. QUERY TOKENIZATION

```python
def tokenize_query(query: str) -> List[str]:
    """
    Splits query into terms, parentheses, and operators.
    Supports AND, OR, NOT, (, )
    """
    pattern = r"\(|\)|AND|OR|NOT|[a-zA-Z]+"
    raw_tokens = re.findall(pattern, query, flags=re.IGNORECASE)

    tokens = []
    for tok in raw_tokens:
        upper = tok.upper()
        if upper in {"AND", "OR", "NOT", "(", ")"}:
            tokens.append(upper)
        else:
            normalized = normalize_text(tok)
            if normalized:
                tokens.extend(normalized)
    return tokens
```

## 6. PARSER (SHUNTING-YARD)

```python
PRECEDENCE = {
    "NOT": 3,
    "AND": 2,
    "OR": 1
}

def to_postfix(tokens: List[str]) -> List[str]:
    """
    Convert infix Boolean query to postfix notation.
    NOT has highest precedence, then AND, then OR.
    """
    output = []
    stack = []

    for token in tokens:
        if token not in {"AND", "OR", "NOT", "(", ")"}:
            output.append(token)

        elif token == "(":
            stack.append(token)

        elif token == ")":
            while stack and stack[-1] != "(":
                output.append(stack.pop())
            if not stack:
                raise ValueError("Mismatched parentheses")
            stack.pop()

        else:
            while (
                stack and stack[-1] != "("
                and PRECEDENCE.get(stack[-1], 0) >= PRECEDENCE[token]
            ):
                output.append(stack.pop())
            stack.append(token)

    while stack:
        if stack[-1] in {"(", ")"}:
            raise ValueError("Mismatched parentheses")
        output.append(stack.pop())

    return output
```

## 7. BOOLEAN EVALUATION


```python
def evaluate_postfix(postfix: List[str], index: Dict[str, Dict[int, int]], universe: Set[int]) -> Set[int]:
    """
    Evaluate Boolean expression in postfix form using set operations.
    """
    stack = []

    for token in postfix:
        if token not in {"AND", "OR", "NOT"}:
            posting_docs = set(index.get(token, {}).keys())
            stack.append(posting_docs)

        elif token == "NOT":
            if len(stack) < 1:
                raise ValueError("Invalid query: NOT missing operand")
            operand = stack.pop()
            stack.append(universe - operand)

        else:
            if len(stack) < 2:
                raise ValueError(f"Invalid query: {token} missing operands")
            right = stack.pop()
            left = stack.pop()

            if token == "AND":
                stack.append(left & right)
            elif token == "OR":
                stack.append(left | right)

    if len(stack) != 1:
        raise ValueError("Invalid query")

    return stack[0]


def evaluate_boolean_query(query: str, index: Dict[str, Dict[int, int]], num_docs: int) -> Set[int]:
    tokens = tokenize_query(query)
    postfix = to_postfix(tokens)
    universe = set(range(num_docs))
    return evaluate_postfix(postfix, index, universe)
```

## 8. INDEX STATISTICS


```python
def compute_index_stats(index: Dict[str, Dict[int, int]]) -> Dict[str, float]:
    num_tokens = len(index)
    posting_lengths = [len(postings) for postings in index.values()]
    total_entries = sum(posting_lengths)
    avg_posting_length = total_entries / num_tokens if num_tokens else 0.0

    return {
        "num_tokens": num_tokens,
        "total_entries": total_entries,
        "avg_posting_length": avg_posting_length
    }
```

## 9. DISPLAY SNIPPETS

```python
def first_sentence(text: str, max_len: int = 200) -> str:
    parts = re.split(r"(?<=[.!?])\s+", text.strip())
    if not parts or not parts[0]:
        snippet = text.strip()[:max_len]
        return snippet
    return parts[0][:max_len]
```

## 10. EXTENDED BOOLEAN MODEL

```python
def document_frequency(index: Dict[str, Dict[int, int]], term: str) -> int:
    return len(index.get(term, {}))


def query_terms_only(tokens: List[str]) -> List[str]:
    return [t for t in tokens if t not in {"AND", "OR", "NOT", "(", ")"}]


def normalized_tf(tf: int) -> float:
    """
    Saturating TF, not TF-IDF.
    Produces values in [0, 1).
    """
    return tf / (tf + 1) if tf > 0 else 0.0


def rarity_weight(df: int, N: int) -> float:
    """
    Not TF-IDF. This is a custom rarity weight in (0,1].
    Rare terms get higher weight.
    """
    if df == 0:
        return 1.0
    return 1.0 - (df / N)


def score_extended_boolean_query(query: str, index, doc_tokens, num_docs: int) -> List[Tuple[int, float]]:
    """
    Custom weighted Boolean scoring approach:
    - Positive terms add score based on:
        saturation_tf * rarity_weight
    - OR contributes if any positive term is present
    - AND favors documents that satisfy more required terms
    - NOT strongly penalizes docs containing negated terms

    This is intentionally NOT tf-idf.
    """
    tokens = tokenize_query(query)

    positive_terms = []
    negated_terms = []

    i = 0
    while i < len(tokens):
        if tokens[i] == "NOT" and i + 1 < len(tokens):
            if tokens[i + 1] not in {"AND", "OR", "NOT", "(", ")"}:
                negated_terms.append(tokens[i + 1])
                i += 2
                continue
        if tokens[i] not in {"AND", "OR", "NOT", "(", ")"}:
            positive_terms.append(tokens[i])
        i += 1

    positive_terms = list(dict.fromkeys(positive_terms))
    negated_terms = list(dict.fromkeys(negated_terms))

    scores = []

    for doc_id, tf_counter in doc_tokens.items():
        pos_score = 0.0
        matched_pos = 0

        for term in positive_terms:
            tf = tf_counter.get(term, 0)
            df = document_frequency(index, term)
            w_tf = normalized_tf(tf)
            w_rarity = rarity_weight(df, num_docs)
            term_score = w_tf * w_rarity
            pos_score += term_score
            if tf > 0:
                matched_pos += 1

        coverage = matched_pos / len(positive_terms) if positive_terms else 0.0

        neg_penalty = 0.0
        for term in negated_terms:
            tf = tf_counter.get(term, 0)
            if tf > 0:
                neg_penalty += normalized_tf(tf)

        final_score = (0.7 * pos_score + 0.3 * coverage) - 0.8 * neg_penalty

        if final_score > 0:
            scores.append((doc_id, final_score))

    scores.sort(key=lambda x: x[1], reverse=True)
    return scores
```

```python
from IPython.display import display, HTML
import html
import math


def highlight_query_terms(text: str, query_terms: list[str]) -> str:
    """
    Very simple highlighter for snippet text.
    Escapes HTML first, then wraps query terms in <b>...</b>.
    """
    safe_text = html.escape(text)

    # sort longer terms first so smaller terms do not break larger ones
    unique_terms = sorted(set(query_terms), key=len, reverse=True)

    for term in unique_terms:
        if not term:
            continue
        # simple case-insensitive replacement
        import re
        pattern = re.compile(rf"\b({re.escape(term)})\b", re.IGNORECASE)
        safe_text = pattern.sub(r"<b>\1</b>", safe_text)

    return safe_text


def extract_query_terms_for_display(query: str) -> list[str]:
    """
    Extract only real terms from the Boolean query for highlighting.
    Requires your tokenize_query() function.
    """
    tokens = tokenize_query(query)
    return [t for t in tokens if t not in {"AND", "OR", "NOT", "(", ")"}]


def render_google_like_results(title: str, query: str, results: list[dict], max_results: int = 10):
    """
    Renders search results in a Google-like style in Jupyter Notebook.
    results = list of dicts with keys:
        - doc_id
        - snippet
        - url (optional fake display URL)
        - score (optional)
    """
    query_terms = extract_query_terms_for_display(query)

    header = f"""
    <div style="font-family: Arial, sans-serif; max-width: 900px; margin: 20px auto;">
        <h2 style="color:#202124; margin-bottom: 6px;">{html.escape(title)}</h2>
        <div style="color:#5f6368; font-size:14px; margin-bottom: 20px;">
            Query: <span style="color:#202124; font-weight:bold;">{html.escape(query)}</span><br>
            About {len(results)} results
        </div>
    """

    body = ""
    for item in results[:max_results]:
        doc_id = item["doc_id"]
        snippet = highlight_query_terms(item["snippet"], query_terms)

        score_html = ""
        if "score" in item and item["score"] is not None:
            score_html = f"""
            <div style="color:#188038; font-size:12px; margin-top:2px;">
                Relevance score: {item['score']:.4f}
            </div>
            """

        fake_url = item.get("url", f"doc://{doc_id}")

        body += f"""
        <div style="margin-bottom: 24px;">
            <div style="font-size:14px; color:#202124; margin-bottom:2px;">
                {html.escape(fake_url)}
            </div>
            <div style="font-size:20px; color:#1a0dab; margin-bottom:4px; cursor:pointer;">
                Document {doc_id}
            </div>
            {score_html}
            <div style="font-size:14px; color:#4d5156; line-height:1.5;">
                {snippet}
            </div>
        </div>
        """

    footer = "</div>"

    display(HTML(header + body + footer))
```

```python
from IPython.display import display, HTML
import html
import math


def highlight_query_terms(text: str, query_terms: list[str]) -> str:
    """
    Very simple highlighter for snippet text.
    Escapes HTML first, then wraps query terms in <b>...</b>.
    """
    safe_text = html.escape(text)

    # sort longer terms first so smaller terms do not break larger ones
    unique_terms = sorted(set(query_terms), key=len, reverse=True)

    for term in unique_terms:
        if not term:
            continue
        # simple case-insensitive replacement
        import re
        pattern = re.compile(rf"\b({re.escape(term)})\b", re.IGNORECASE)
        safe_text = pattern.sub(r"<b>\1</b>", safe_text)

    return safe_text


def extract_query_terms_for_display(query: str) -> list[str]:
    """
    Extract only real terms from the Boolean query for highlighting.
    Requires your tokenize_query() function.
    """
    tokens = tokenize_query(query)
    return [t for t in tokens if t not in {"AND", "OR", "NOT", "(", ")"}]


def render_google_like_results(title: str, query: str, results: list[dict], max_results: int = 10):
    """
    Renders search results in a Google-like style in Jupyter Notebook.
    results = list of dicts with keys:
        - doc_id
        - snippet
        - url (optional fake display URL)
        - score (optional)
    """
    query_terms = extract_query_terms_for_display(query)

    header = f"""
    <div style="font-family: Arial, sans-serif; max-width: 900px; margin: 20px auto;">
        <h2 style="color:#202124; margin-bottom: 6px;">{html.escape(title)}</h2>
        <div style="color:#5f6368; font-size:14px; margin-bottom: 20px;">
            Query: <span style="color:#202124; font-weight:bold;">{html.escape(query)}</span><br>
            About {len(results)} results
        </div>
    """

    body = ""
    for item in results[:max_results]:
        doc_id = item["doc_id"]
        snippet = highlight_query_terms(item["snippet"], query_terms)

        score_html = ""
        if "score" in item and item["score"] is not None:
            score_html = f"""
            <div style="color:#188038; font-size:12px; margin-top:2px;">
                Relevance score: {item['score']:.4f}
            </div>
            """

        fake_url = item.get("url", f"doc://{doc_id}")

        body += f"""
        <div style="margin-bottom: 24px;">
            <div style="font-size:14px; color:#202124; margin-bottom:2px;">
                {html.escape(fake_url)}
            </div>
            <div style="font-size:20px; color:#1a0dab; margin-bottom:4px; cursor:pointer;">
                Document {doc_id}
            </div>
            {score_html}
            <div style="font-size:14px; color:#4d5156; line-height:1.5;">
                {snippet}
            </div>
        </div>
        """

    footer = "</div>"

    display(HTML(header + body + footer))
```

```python
def prepare_ranked_results(query: str, documents, index, doc_tokens, max_results: int = 10):
    """
    Extended Boolean retrieval:
    returns a ranked list of dicts for display.
    """
    ranked = score_extended_boolean_query(query, index, doc_tokens, len(documents))

    results = []
    for doc_id, score in ranked[:max_results]:
        snippet = first_sentence(documents[doc_id], max_len=220)
        results.append({
            "doc_id": doc_id,
            "snippet": snippet,
            "url": f"doc://ranked/{doc_id}",
            "score": score
        })

    return results, len(ranked)
```

```python
def search_both_methods(query: str, documents, index, doc_tokens, max_results: int = 10):
    """
    Runs both:
    1. Standard Boolean retrieval
    2. Extended Boolean ranked retrieval

    Displays both in a Google-like format inside Jupyter.
    """
    try:
        boolean_results, boolean_count = prepare_boolean_results(
            query=query,
            documents=documents,
            index=index,
            max_results=max_results
        )

        ranked_results, ranked_count = prepare_ranked_results(
            query=query,
            documents=documents,
            index=index,
            doc_tokens=doc_tokens,
            max_results=max_results
        )

        display(HTML("""
        <hr style="margin:30px 0;">
        <h1 style="font-family:Arial, sans-serif; color:#202124; text-align:center;">
            Search Comparison
        </h1>
        """))

        render_google_like_results(
            title=f"Standard Boolean Retrieval ({boolean_count} matches)",
            query=query,
            results=boolean_results,
            max_results=max_results
        )

        display(HTML("<hr style='margin:30px 0;'>"))

        render_google_like_results(
            title=f"Extended Boolean Ranked Retrieval ({ranked_count} ranked documents)",
            query=query,
            results=ranked_results,
            max_results=max_results
        )

    except Exception as e:
        display(HTML(f"""
        <div style="font-family:Arial, sans-serif; color:#b00020; font-size:16px;">
            <b>Error:</b> {html.escape(str(e))}
        </div>
        """))
```

```python
import ipywidgets as widgets
from IPython.display import display


def boolean_search_interface(documents, index, doc_tokens, max_results: int = 10):
    """
    Interactive search box for Jupyter Notebook.
    """
    query_box = widgets.Text(
        value="space AND nasa",
        placeholder="Enter Boolean query...",
        description="Query:",
        layout=widgets.Layout(width="70%")
    )

    search_button = widgets.Button(
        description="Search",
        button_style="primary"
    )

    output = widgets.Output()

    def on_search_click(_):
        with output:
            output.clear_output()
            search_both_methods(
                query=query_box.value,
                documents=documents,
                index=index,
                doc_tokens=doc_tokens,
                max_results=max_results
            )

    search_button.on_click(on_search_click)

    display(widgets.HBox([query_box, search_button]))
    display(output)
```

```python
from IPython.display import display, HTML
import html
import re

# ----------------------------
# Small helpers
# ----------------------------

def first_sentence(text: str, max_len: int = 220) -> str:
    text = " ".join(text.strip().split())
    if not text:
        return "[No preview available]"
    parts = re.split(r'(?<=[.!?])\s+', text)
    snippet = parts[0] if parts and parts[0] else text
    return snippet[:max_len] + ("..." if len(snippet) > max_len else "")


def extract_query_terms_for_display(query: str) -> list[str]:
    tokens = tokenize_query(query)
    return [t for t in tokens if t not in {"AND", "OR", "NOT", "(", ")"}]


def highlight_query_terms(text: str, query_terms: list[str]) -> str:
    safe_text = html.escape(text)
    unique_terms = sorted(set(query_terms), key=len, reverse=True)

    for term in unique_terms:
        if not term:
            continue
        pattern = re.compile(rf"\b({re.escape(term)})\b", re.IGNORECASE)
        safe_text = pattern.sub(r"<b>\1</b>", safe_text)

    return safe_text


# ----------------------------
# Result rendering
# ----------------------------

def render_google_like_results(title: str, query: str, results: list[dict], total_count: int):
    query_terms = extract_query_terms_for_display(query)

    html_output = f"""
    <div style="font-family: Arial, sans-serif; max-width: 900px; margin: 20px auto;background-color:#fff;">
        <h2 style="color:#202124; margin-bottom: 8px;">{html.escape(title)}</h2>
        <div style="color:#5f6368; font-size:14px; margin-bottom: 20px;">
            Query: <b>{html.escape(query)}</b><br>
            About {total_count} results
        </div>
    """

    if not results:
        html_output += """
        <div style="font-size:14px; color:#4d5156; margin-bottom:20px;">
            No matching documents found.
        </div>
        """
    else:
        for item in results:
            doc_id = item["doc_id"]
            snippet = highlight_query_terms(item["snippet"], query_terms)
            fake_url = item.get("url", f"doc://{doc_id}")

            score_html = ""
            if "score" in item and item["score"] is not None:
                score_html = f"""
                <div style="color:#188038; font-size:12px; margin-top:2px;">
                    Relevance score: {item['score']:.4f}
                </div>
                """

            html_output += f"""
            <div style="margin-bottom: 24px;">
                <div style="font-size:14px; color:#202124; margin-bottom:2px;">
                    {html.escape(fake_url)}
                </div>
                <div style="font-size:20px; color:#1a0dab; margin-bottom:4px;">
                    Document {doc_id}
                </div>
                {score_html}
                <div style="font-size:14px; color:#4d5156; line-height:1.5;">
                    {snippet}
                </div>
            </div>
            """

    html_output += "</div>"
    display(HTML(html_output))


# ----------------------------
# Prepare results for BOTH methods
# ----------------------------

def prepare_boolean_results(query: str, documents, index, max_results: int = 10):
    matched_doc_ids = evaluate_boolean_query(query, index, len(documents))
    matched_doc_ids = sorted(matched_doc_ids)

    results = []
    for doc_id in matched_doc_ids[:max_results]:
        results.append({
            "doc_id": doc_id,
            "snippet": first_sentence(documents[doc_id]),
            "url": f"doc://boolean/{doc_id}"
        })

    return results, len(matched_doc_ids)


def prepare_ranked_results(query: str, documents, index, doc_tokens, max_results: int = 10):
    ranked = score_extended_boolean_query(query, index, doc_tokens, len(documents))

    results = []
    for doc_id, score in ranked[:max_results]:
        results.append({
            "doc_id": doc_id,
            "snippet": first_sentence(documents[doc_id]),
            "url": f"doc://ranked/{doc_id}",
            "score": score
        })

    return results, len(ranked)


# ----------------------------
# Main comparison function
# ----------------------------

def search_both_methods(query: str, documents, index, doc_tokens, max_results: int = 10):
    try:
        boolean_results, boolean_count = prepare_boolean_results(
            query, documents, index, max_results=max_results
        )

        ranked_results, ranked_count = prepare_ranked_results(
            query, documents, index, doc_tokens, max_results=max_results
        )

        display(HTML("<hr>"))
        render_google_like_results(
            "Standard Boolean Retrieval",
            query,
            boolean_results,
            boolean_count
        )

        display(HTML("<hr>"))
        render_google_like_results(
            "Extended Boolean Ranked Retrieval",
            query,
            ranked_results,
            ranked_count
        )

    except Exception as e:
        display(HTML(f"""
        <div style="color:#b00020; font-family:Arial, sans-serif;">
            <b>Error:</b> {html.escape(str(e))}
        </div>
        """))
```

```python
def run_demo():
    documents = load_corpus(limit=2000)
    index, doc_tokens, normalized_docs = build_inverted_index(documents)

    stats = compute_index_stats(index)
    print("INDEX STATISTICS")
    print(f"Number of unique tokens: {stats['num_tokens']}")
    print(f"Total posting entries: {stats['total_entries']}")
    print(f"Average posting list length: {stats['avg_posting_length']:.2f}")
    print()

    sample_queries = [
        "space AND nasa",
        "graphics OR image",
        "windows AND NOT microsoft",
        "(space OR nasa) AND shuttle",
        "(graphics OR image) AND NOT windows"
    ]

    print("BOOLEAN QUERY EXAMPLES")
    for q in sample_queries:
        results = evaluate_boolean_query(q, index, len(documents))
        print(f"\nQuery: {q}")
        print(f"Matched documents: {len(results)}")
        for doc_id in list(sorted(results))[:5]:
            print(f"  Doc {doc_id}: {first_sentence(documents[doc_id])}")

    print("\nEXTENDED BOOLEAN RANKED EXAMPLE")
    q = "(space OR nasa) AND shuttle NOT moon"
    ranked = score_extended_boolean_query(q, index, doc_tokens, len(documents))
    print(f"Query: {q}")
    for doc_id, score in ranked[:10]:
        print(f"  Doc {doc_id} | score={score:.4f} | {first_sentence(documents[doc_id])}")

    print()
    


if __name__ == "__main__":
    run_demo()
```
